```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_gyro = pd.read_csv('data/car/2018-08-17_16-45-30/Gyroscope.csv')
df_acc = pd.read_csv('data/car/2018-08-17_16-45-30/AccelerometerLinear.csv')
print("Acc.  shape -", df_acc.shape)
print("Gyro. shape -", df_gyro.shape)
print("Gyro columns", df_gyro.columns)
print("Acc  columns", df_acc.columns)

```

```python
df = df_acc
df.rename(columns={'X': 'AX', 'Y': 'AY', 'Z':'AZ'}, inplace=True)
df['GX'] = df_gyro['X']
df['GY'] = df_gyro['Y']
df['GZ'] = df_gyro['Z']
```

```python
print("shape -", df.shape)
print("columns", df.columns)
df.head()
```

```python
df= df[500:4000]
x_gyro = df['GX']
y_gyro = df['GY']
z_gyro = df['GZ']

x_acc = df['AX']
y_acc = df['AY']
z_acc = df['AZ']
```

```python
x_gyro.plot(figsize=(20,4), title="x-axis of gyro for car")
```

```python
y_gyro.plot(figsize=(20,4), title="y-axis of gyro for car")
```

```python
z_gyro.plot(figsize=(20,4), title="z-axis of gyro for car")
```

```python
x_acc.plot(figsize=(20,4), title="x-axis of acc for car")
```

```python
y_acc.plot(figsize=(20,4), title="y-axis of acc for car")
```

```python
z_acc.plot(figsize=(20,4), title="z-axis of acc for car")
```
