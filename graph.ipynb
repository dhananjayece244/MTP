```python
#read input csv file

acc_df = pd.read_csv('AccelerometerLinear.csv')
acc_df = acc_df[500:80000]
acc_df.columns
```

```python
time = acc_df['Milliseconds']
xAcc = acc_df['X']
yAcc = acc_df['Y']
zAcc = acc_df['Z']

# xG = acc_df['gFx']
# yG = acc_df['gFy']
# zG = acc_df['gFz']

# xGyro = acc_df['wx']
# yGyro = acc_df['wy']
# zGyro = acc_df['wz']

# xGforce = acc_df['gFx']

ground_truth = acc_df['ground truth']

# acc_df['Magnitude'] = (xAcc**2 + yAcc**2 + zAcc**2)**(1/2)
```

```python
#line graph
#plt.plot(time, xAcc, label = 'X-axis of Accelerometer')
#plt.plot(time, yAcc, label = 'Y-axis of Accelerometer')
#plt.plot(time, zAcc, label = 'Z-axis of Accelerometer')
plt.plot(time, zAcc, label = 'total sum', color ='g')
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.title('Accelerometer data')
plt.legend()
plt.show()
```

```python
# Function to map the colors as a list from the input list of x variables
def pltcolor(lst):
    cols=[]
    colors = {'RB':'r', 'LR':'b','Lturn':'green', 'Rturn':'yellow','acc':'magenta',
             'dec':'cyan',nan:'black'}
    for l in lst:
        cols.append(colors[l])
        
    return cols
#Create the colors list using the function above
cols=pltcolor(ground_truth)

```

```python
from pylab import *
points = list(zip(time,zAcc ))
segments = list(zip(points[:-1], points[1:]))
plt.figure(figsize=(40,10))
ax = axes(frameon=True)
# ax.legend(("red","Blue"), ('Nothing', 'acc'), title="Legend", loc='upper right', shadow=True)
LC = LineCollection(segments, colors = cols,  label = 'total sum')
ax.add_collection(LC)
axis([0, 56960 , -4, +4])
#savefig('colored_line.png', dpi=70)
show()
```
